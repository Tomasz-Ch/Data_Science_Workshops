```python
import matplotlib.pyplot as plt
from collections import Counter
```

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]
```

```python
plt.plot(years, gdp, color='red', marker='o', linestyle='solid')
plt.title("Nominalny PKB")
plt.ylabel('Mld dolarów')
plt.xlabel("Lata")
plt.show()
```

```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]
```

```python
xs = [i for i, _ in enumerate(movies)]
```

```python
plt.bar(xs, num_oscars, 0.7)
plt.ylabel("Liczba nagród")
plt.title('Moje ulubione filmy')
plt.xticks(xs, movies)
plt.show()
```

```python
grades = [83,95,91,87,70,0,85,82,100,67,73,77,0]
histogram = Counter(decile(grade) for grade in grades)
histogram
```

```python
plt.bar([x for x in histogram.keys()], histogram.values(), 8)
plt.title('Rozkład ocen')
plt.ylabel('Liczba studentów')
plt.xlabel('Decyl')
plt.axis([-5, 105, 0, 5])
plt.show()
```
