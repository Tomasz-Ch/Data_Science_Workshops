```python
import matplotlib.pyplot as plt
from collections import Counter
```

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]
```

```python
plt.plot(years, gdp, color='red', marker='o', linestyle='solid')
plt.title("Nominalny PKB")
plt.ylabel('Mld dolarów')
plt.xlabel("Lata")
plt.grid()
plt.show()
```

```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]
```

```python
xs = [i for i, _ in enumerate(movies)]
```

```python
plt.bar(xs, num_oscars, 0.7)
plt.ylabel("Liczba nagród")
plt.title('Moje ulubione filmy')
plt.xticks(xs, movies)
plt.show()
```

```python
grades = [83,95,91,87,70,0,85,82,100,67,73,77,0]
decile = lambda grade: grade // 10 * 10
histogram = Counter(decile(grade) for grade in grades)
histogram
```

```python
plt.bar([x for x in histogram.keys()], histogram.values(), 8)
plt.title('Rozkład ocen')
plt.ylabel('Liczba studentów')
plt.xlabel('Decyl')
plt.axis([-5, 105, 0, 5])
plt.xticks([10 * i for i in range(11)])
plt.show()
```

```python
variance = [1,2,4,8,16,32,64,128,256]
bias_squared = [256,128,64,32,16,8,4,2,1]
```

```python
total_error = [x + y for x, y in zip(variance, bias_squared)]
xs = range(len(variance))
```

```python
plt.plot(xs, variance, 'g-', label='variance')
plt.plot(xs, bias_squared, 'r-.', label='$bias^2$')
plt.plot(xs, total_error, 'b', label='total_error')
plt.legend(loc=9, frameon=False)
plt.title('Kompromis pomiędzy wartością progową i złożonością modelu')
plt.xlabel('Stopień skomplikowania modelu')
plt.show()
```

```python
friends = [ 70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
```

```python
plt.scatter(friends, minutes)
for label, friend_count,minute_count in zip(labels, friends, minutes):
    plt.annotate(label, xy=(friend_count, minute_count), 
                 xytext=(5, -5), 
                 textcoords='offset points')
plt.title('czas spędzony na stronie a liczba znajomych')
plt.xlabel('Liczba znajomych')
plt.ylabel('Dzienny czas spędzony na stronie (minuty)')
plt.show()
```
